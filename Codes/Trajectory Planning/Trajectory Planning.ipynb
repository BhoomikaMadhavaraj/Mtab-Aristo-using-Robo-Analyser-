```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Tranjectory_planning(theta_initial, theta_final, time):
    t = np.arange(0,time,0.1)  
    a_1 = 0

    for i in range(len(theta_initial)):
        POS = []
        VEL = []
        ACC = []
        a_0 = theta_initial[i]
        for tf in t:
            a_2 = (3/tf**2)*(theta_final[i] - theta_initial[i])
            a_3 = (-2/tf**3)*(theta_final[i] - theta_initial[i])
            position = a_0 + a_1*tf + a_2*(tf**2) + a_3*(tf**3)
            velocity = a_1 + 2*a_2*tf + 3*a_3*(tf**2)
            acceleration = 2*a_2 + 6*a_3*tf
            POS.append(position)
            VEL.append(velocity)
            ACC.append(acceleration)
        plt.plot(t,POS)
        plt.xlabel('Time')
        plt.ylabel('Position')
        plt.title('POSITIION VS TIME PLOT OF ARM')
        plt.show()
        plt.plot(t, VEL)
        plt.xlabel('Time')
        plt.ylabel('Velocity')
        plt.title('VELOCITY VS TIME PLOT OF ARM')
        plt.show()
        plt.plot(t, ACC)
        plt.xlabel('Time')
        plt.ylabel('Acceleration')
        plt.title('ACCELERATION VS TIME PLOT OF ARM')
        plt.show()
```

```python
theta_initial = [15, 21, 30, 45, 75, 20]
theta_final = [75, 30, 45, 20, 100, 60]
time = 3

Tranjectory_planning(theta_initial, theta_final, time)
```
